```python
# Importação das bibliotecas
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
# Exibir gráficos inline
%matplotlib inline

# Carregar imagem de exemplo
from PIL import Image
import requests
url = "https://images.unsplash.com/photo-1745810187217-4d9e1ccfd9d5?ixlib=rb-4.1.0&q=85&fm=jpg&crop=entropy&cs=srgb&w=640"
image = Image.open(requests.get(url, stream=True).raw)
image = image.convert("RGB")
# Mostrar imagem original
plt.imshow(image)
plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

```python
# 1- EXIBIR APENAS O CANAL VERMELHO DA IMAGEM

# Separar os canais
r, g, b = image.split()

# Exibir apenas o canal vermelho da imagem
plt.imshow(r, cmap='gray')
plt.title("Imagem com o Canal Vermelho")
plt.axis('off')
plt.show()
```

```python
# 2- ALTERAR O CANAL VERMELHO DA IMAGEM

# Converter canal vermelho para numpy
arrayRed = np.array(r)

# Valores de cada pixel do canal vermelho recebem +20
arrayRed += 20

# Recria a imagem com o canal vermelho alterado
novaImagem = Image.merge("RGB", (Image.fromarray(arrayRed),g,b))

# Exibir imagem alterada
plt.imshow(novaImagem)
plt.title("Imagem Original com o Canal Vermelho Modificado")
plt.axis('off')
plt.show()
```

```python
# 3- CRIAR UMA IMAGEM NEGATIVA (INVERTER OS VALORES)

np_img = np.array(image)

# Inverter as cores
imagemNegativa = 255 - np_img
"""
255 - [Valor de cada pixel da imagem original] resulta no negativo,

Exemplo: pixel 0 se torna 255 e pixel 255 se torna 0,

[255 - 0 da imagemOriginal = 255] && [255 - 255 da imagemOriginal = 0].
"""

# Exibir imagem negativa
plt.imshow(imagemNegativa)
plt.title("Imagem Negativa")
plt.axis('off')
plt.show()
```

```python
# 4- SALVAR IMAGEM COMO ARQUIVO

arrayImgNgtv = np.array(imagemNegativa)

# Salvar imagem negativa como arquivo
imgNegativa = Image.fromarray(arrayImgNgtv)
imgNegativa.save("imagemNegativa.jpg")

# Baixar arquivo gerado no Colab
from google.colab import files
files.download("imagemNegativa.jpg")
```

```python
# 5- RECORTE UM PEDÇAO DA IMAGEM E SALVE EM DISCO

# Converter imagem para array numpy
np_image = np.array(image)

# Obter altura e largura
altura,largura, _ = np_image.shape

# Definir recorte do canto superior direito (metade da largura e altura)
corte = np_image[0 : altura//2, largura//2 : largura]
"""
Corta a imagem do pixel 0 até metade da altura e do pixel do meio largura até o final.
"""

# Converter de volta para imagem PIL
imagemCortada = Image.fromarray(corte)

# Exibir imagem recortada
plt.imshow(imagemCortada)
plt.title("Canto Superior Direito Recortado")
plt.axis('off')
plt.show()

# Salvar e baixar
imagemCortada.save("imagemCortada.jpg")
files.download("imagemCortada.jpg")
```
